# Scraping the Wikipedia Page for the Postal Codes

```python
source = requests.get('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=945633050').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

## Converting HTML table to Pandas DataFrame

```python
DFS = pd.read_html(tab)
DF=DFS[0]
DF.head()
```

# Data Preprocessing and Cleaning

```python
# Dropping the rows where Borough = 'Not Assigned'

df1 = DF[DF.Borough != 'Not Assigned']

# Combining the Neighbourhoods with the same Postal Code

df2 = df1.groupby(['Postcode', 'Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

#Replacing the name of the Neighbourhoods which = 'Not Assigned' with the Borough Names

df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned', df2['Borough'], df2['Neighbourhood'])

df2
```

## Shape of Data Frame

```python
df2.shape
```

# Importing the Longitude and Latitude CSV file for various Neighbourhoods

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

## Merging the Latitude and Longtiude Tables

```python
lat_lon.rename(columns={'Postal Code':'Postcode'}, inplace=True)
df3 = pd.merge(df2,lat_lon,on='Postcode')
df3.head()
```

# The notebook from here will include the Clustering and the Plotting of the Neighbourhoods which contain Toronto in their Borough

### Gathering all the rows from the dataframe that have Toronto in their Boroughs

```python
df4 = df3[df3['Borough'].str.contains('Toronto', regex=False)]
df4
```

### Visualizing all the Neighbourhoods with Folium using the above dataframe

```python
Toronto_map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat, lng, borough, neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='green',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(Toronto_map)
Toronto_map
```

# Using KMeans for clustering the Neighbourhoods

```python
k=5
Toronto_clustering = df4.drop(['Postcode', 'Borough', 'Neighbourhood'],1)
kmeans = KMeans(n_clusters = k, random_state=0).fit(Toronto_clustering)
kmeans.labels_
```

```python
# Creating a map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# Set Color Scheme
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range (k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# Adding Markers
markers_colors = []
for lat, lon, nieghbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
    [lat, lon],
    radius=5,
    popup=label,
    color=rainbow[cluster-1],
    fill=True,
    fill_color=rainbow[cluster-1],
    fill_opacity=0.7).add_to(map_clusters)
    
map_clusters
```
